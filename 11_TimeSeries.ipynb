```python
import plotly.graph_objects as go
from plotly import offline
from datetime import datetime
import pandas_datareader.data as web
import pandas as pd

pd.core.common.is_list_like = pd.api.types.is_list_like

offline.init_notebook_mode(connected=True)
```

```python
# read amazon stock data from stooq website
amzn = web.DataReader('AMZN', 'stooq',
                      datetime(2017, 1, 1),
                      datetime(2018, 6, 1)
                      )
amzn.head()
```

```python
amzn.shape
```

```python
amzn.reset_index(inplace=True)
amzn.head()
```

```python
trace = go.Scatter(
    x=amzn.Date,
    y=amzn.High,
)
data = [trace]
offline.iplot(data)
```

```python
trace_high = go.Scatter(
    x=amzn.Date,
    y=amzn.High,

    name='Amazon High',
    line=go.scatter.Line(
        color='#6699FF'
    )
)

trace_low = go.Scatter(
    x=amzn.Date,
    y=amzn.Low,

    name='Amazon Low',
    line=go.scatter.Line(
        color='#FF6633'
    )
)

data = [trace_high, trace_low]
layout = go.Layout(
    title='Amazon Stock Price Data'
)
figure = go.Figure(data=data, layout=layout)
offline.iplot(figure)
```

```python
layout = go.Layout(
    width=1000,
    height=600,
    title='Amazon Stock Price Data',
    xaxis=go.layout.XAxis(
        rangeslider=go.layout.xaxis.Rangeslider(
            autorange=True
        ),
        type='date'
    )
)
figure = go.Figure(data=data, layout=layout)
offline.iplot(figure)
```

```python
layout = go.Layout(
    width=1000,
    height=600,
    title='Amazon Stock Price Data',
    xaxis=go.layout.XAxis(
        rangeslider=go.layout.xaxis.Rangeslider(
            autorange=True
        ),
        type='date',
        rangeselector=go.layout.xaxis.Rangeselector(
            buttons=[
                go.layout.xaxis.rangeselector.Button(
                    count=1, label='1m', step='month', stepmode='todate',
                ),
                go.layout.xaxis.rangeselector.Button(
                    count=6, label='6m', step='month', stepmode='backward'
                ),
                go.layout.xaxis.rangeselector.Button(
                    step='all',
                )
            ]
        )
    )
)
figure = go.Figure(data=data, layout=layout)
offline.iplot(figure)
```
