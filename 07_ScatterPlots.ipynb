```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from plotly import offline

offline.init_notebook_mode(connected=True)
```

```python
n = 200
random_x = np.random.randn(n)
random_y = np.random.randn(n)

trace = go.Scatter(
    x=random_x,
    y=random_y,
    mode='markers'
)

data = [trace]

offline.iplot(data)
```

```python
n = 50
random_x = np.linspace(0, 1, n)

random_y0 = np.random.randn(n)+5
random_y1 = np.random.randn(n)
random_y2 = np.random.randn(n)-5
```

```python
trace0 = go.Scatter(
    x=random_x,
    y=random_y0,
    mode='markers'
)
trace1 = go.Scatter(
    x=random_x,
    y=random_y1,
    mode='lines'
)
trace2 = go.Scatter(
    x=random_x,
    y=random_y2,
    mode='lines+markers'
)

data = [trace0, trace1, trace2]
offline.iplot(data)
```

```python
honey_data = pd.read_csv('datasets/honeyproduction.csv')
honey_data.head()
```

```python
trace = go.Scatter(
    x=honey_data['state'],
    y=honey_data['year'],

    mode='markers',
    marker=go.scatter.Marker(
        size=14,
        color=honey_data['totalprod'],
        colorscale='Rainbow',
        showscale=True,
        opacity=0.5
    ),
    text=honey_data['totalprod']
)
data = [trace]
layout = go.Layout(
    title='Hone Production in the United States',
    hovermode='closest',
    xaxis=go.layout.XAxis(
        title='States',
    ),
    yaxis=go.layout.YAxis(
        title='Honey Production'
    )
)

fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```
