```python
import plotly.graph_objects as go
from plotly import offline
import numpy as np
import pandas as pd

offline.init_notebook_mode(connected=True)
```

```python
n = 50
random_x = np.random.randn(n)
random_y = np.random.randn(n)
random_z = np.random.randn(n)
random_a = np.random.randn(n)
random_b = np.random.randn(n)
```

```python
trace = go.Scatter3d(
    x=random_x,
    y=random_y,
    z=random_z,

    mode='markers',
    marker=go.scatter3d.Marker(
        color=random_a,
        colorscale='Rainbow',
        sizemode='area',
        size=1000 * abs(random_b),
        showscale=True,
    ),
)

data = [trace]
layout = go.Layout(
    title='Random 3D Scatter Plot'
)
figure = go.Figure(data=data, layout=layout)
offline.iplot(figure)
```

```python
planets = pd.read_csv('datasets/planets.csv')
planets
```

```python
trace_3d = go.Scatter3d(
    x=planets['planet'],
    y=planets['distance_from_sun'],
    z=planets['no_of_moons'],

    mode='markers',

    marker=go.scatter3d.Marker(
        sizeref=1000,
        size=planets['planet_diameter'],
        color=planets['period_of_revolution_around_sun'],
        colorscale='Rainbow',
        showscale=True,
    )
)
scatter_data = [trace_3d]
layout = go.Layout(
    title='Our solar system'
)
figure = go.Figure(data=scatter_data, layout=layout)
offline.iplot(figure)
```

```python
colors = [
    '#D5D2D1',
    '#F8B581',
    '#182A61',
    '#B53B03',
    '#B9CBD5',
    '#E0CDAD',
    '#D3F9FA',
    '#3454DF',
    '#E9E8D2',
]
trace = go.Scatter3d(
    x=planets['distance_from_sun'],
    y=planets['period_of_revolution_around_sun'],
    z=planets['no_of_moons'],

    text=planets['planet'],
    mode='markers',
    marker=go.scatter3d.Marker(
        sizemode='diameter',
        sizeref=750,
        size=planets['planet_diameter'],
        color=colors
    )
)
data = [trace]
layout = go.Layout(
    width=900,
    height=900,
    title='Planets of our Solar System',
    scene=go.layout.Scene(
        xaxis=go.layout.scene.XAxis(
            title='Distance from Sun',
            color='white',
            titlefont=dict(color='yellow')
        ),
        yaxis=go.layout.scene.YAxis(
            title='Revolution Period (Earth days)',
            color='white',
            titlefont=dict(color='yellow')
        ),
        zaxis=go.layout.scene.ZAxis(
            title='Moon Count',
            color='white',
            titlefont=dict(color='yellow'),
        ),
        bgcolor='#141836',
    )
)
figure = go.Figure(
    data=data, layout=layout
)
offline.iplot(figure)
```
