```python
import plotly.graph_objects as go
from plotly import offline
import pandas as pd

offline.init_notebook_mode(connected=True)
```

```python
x0 = [0]
y0 = [11]
```

```python
x1 = [0, 1]
y1 = [11, 13]
```

```python
x2 = [0, 1, 1]
y2 = [11, 13, 12]
x3 = [2, 3]
y3 = [13, 14]
```

```python
layout = go.Layout(
    xaxis=dict(range=[0, 3]),
    yaxis=dict(range=[10, 15]),

    updatemenus=[
        go.layout.Updatemenu(
            type='buttons',
            buttons=[
                go.layout.updatemenu.Button(
                    label='Play',
                    method='animate',
                    args=[None]
                )
            ]
        )
    ]
)

frames = [
    go.Frame(data=[dict(x=x0, y=y0)]),
    go.Frame(data=[dict(x=x1, y=y1)]),
    go.Frame(data=[dict(x=x2, y=y2)]),
    go.Frame(data=[dict(x=x3, y=y3)]),
]

fig = go.Figure(
    data=[{}],
    layout=layout,
    frames=frames
)
offline.iplot(fig)
```

```python
data = pd.read_csv('datasets/avocado.csv')
data.head()
```

```python
data_ca = data.loc[(data['region'] == 'California') & (data['type'] == 'organic')]
data_ca = data_ca.sort_values(by=['Date'])
data_ca.head()
```

```python
avg_prices = list(data_ca['AveragePrice'])
date_list = list(data_ca['Date'])
```

```python
x_list = []
y_list = []
frames_list = []
```

```python
for i in range(len(date_list)):
    x_list.append(date_list[i])
    y_list.append(avg_prices[i])
    frames_list.append(go.Frame(data=[dict(x=x_list.copy(), y=y_list.copy())]))

```

```python
play_button = go.layout.updatemenu.Button(
    label='Play',
    method='animate',
    args=[None,
          dict(fromcurrent=True,
               transition=dict(duration=200),
               frame=dict(duration=100)
               )
          ]
)
```

```python
pause_button = go.layout.updatemenu.Button(
    label='Pause',
    method='animate',
    args=[[None], dict(mode='immediate')]
)
```

```python
print(date_list[0], '-', date_list[-1])
# date_list[-1]
pd.to_datetime(date_list[0])
```

```python
layout = go.Layout(
    xaxis=go.layout.XAxis(
        type='date',
        range=[date_list[0], date_list[-1]]
    ),
    yaxis=dict(range=[0, 1 + max(avg_prices)]),
    updatemenus=[
        dict(type='buttons',
             buttons=[play_button, pause_button])
    ]
)

fig = go.Figure(data=[{}], layout=layout, frames=frames_list)
offline.iplot(fig)
```
