```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from plotly import offline

offline.init_notebook_mode(connected=True)
```

```python
random_numbers = np.random.randn(200)
```

```python
hist_trace = go.Histogram(
    x=random_numbers
)

data = [hist_trace]
layout = go.Layout(
    xaxis=dict(title='Randomly generated values'),
    yaxis=dict(title='Count')
)
fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```

```python
hist_trace = go.Histogram(
    x=random_numbers,
    histnorm='probability',
)

data = [hist_trace]
layout = go.Layout(
    xaxis=dict(title='Randomly generated values'),
    yaxis=dict(title='Probability')
)
fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```

```python
hist_trace = go.Histogram(
    x=random_numbers,
    cumulative=dict(enabled=True),
)

data = [hist_trace]
layout = go.Layout(
    xaxis=dict(title='Randomly generated values'),
    yaxis=dict(title='Probability')
)
fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```

```python
more_random_numbers = 1 + np.random.randn(100)
```

```python
trace0 = go.Histogram(
    x=random_numbers,
    opacity=0.9
)

trace1 = go.Histogram(
    x=more_random_numbers,
    opacity=0.5
)
data = [trace0, trace1]

layout = go.Layout(
    barmode='overlay'
)
fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```

```python
trace0 = go.Histogram(x=random_numbers,
                      name='Series 1',
                      xbins=dict(start=-4, end=4, size=0.5),
                      marker=dict(color='#D67D93'),
                      opacity=0.75
                      )

trace1 = go.Histogram(x=more_random_numbers,
                      name='Series 2',
                      xbins=dict(start=0, end=4, size=0.5),
                      marker=dict(color='#7daed6'),
                      opacity=0.75
                      )

data = [trace0, trace1]
layout = go.Layout(title='Styled Histogram',
                   xaxis=dict(title='Random Number Value'),
                   yaxis=dict(title='Value Counts'),
                   bargap=0.2,
                   bargroupgap=0.1
                   )
fig = go.Figure(data=data, layout=layout)
offline.iplot(fig)
```
